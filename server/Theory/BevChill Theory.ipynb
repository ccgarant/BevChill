## Problem

It is unknown how long it will take my room temperature beverage, if put the beverage in the freezer or refridgerator, to chill down to the 'perfect' drinking temperature, or at the very least a 'good enough' drinking temperture.

Example: It's Friday, on your way home from work you stop at the beer store and pick out a package of beer.  It's room temperature.  When you arrive home, you throw a few beers in the freezer.  About how long should you wait until they are 'cold enough' to grab.  About how long should you wait until they are 'perfect' to drink.  When will they start to freezer?

## Goal

To develope a 'beverage chill timer' that approximates the beverage's tempertaure as it is cooling in a freezer or refridgerator.

Based on real physics and engineering of heat transfer theory, estimate how long a certain beverage type needs to chill before reaching its 'ideal drinking temperature'.  This approximation shall take into account various starting initial condition.  Verify the approximation predicts reality with emperical testing.

## Approach

From Heat Transfer Theory Ref [1], the approach of __*Thermal Resistance*__ has been chosen.  This approach is a conservation of energy approach that models the energy lost from the closed system as a resistance.  That is, if the heat transfers well thru say metal, the resistance of retaining heat energy is low.  In contrast, glass does not dissipate heat as well as metal, and therefore has a higher thermal resistance than metal.


## Conservation of Energy and Thermal Resistance Approach 

If the beer bottle is treated as a closed system from the initial conditions of being in equilibrium with the atmospheric temperature Tatm , and treating t0 as the starting time once placed in the refrigerator, the thermal mass has a stored energy which will be taken away by the cooler refrigerator conditions.  Given the conservation of energy

$$
-\dot{E}_{out}= \dot{E}_{stored} \\
$$

The energy stored is a function of the material physical properties of the bottle and beverage, and the change in temperature with time, here only using the beverage properties as they are largely dominant.

$$
\dot{E}_{stored}= \dot{q}_{stored}=\rho V c\frac{d\mathbf{T}}{d\mathbf{t}}\\
$$

The energy dissipated can be modelled as *Thermal Resistance*, each heat transfer mode having its own resistance to dissipating heat, much like electrical resistance.  The Thermal Resistance approach is given by

$$
-\dot{E}_{out}= Q = U A (T_{2,\infty}-T_{center}) \\
$$

Where U is the overall heat transfer coefficient, and A is the surface area.  

For our application, a bottled beer being cooled can have a thermal resistance of internal convection between the beer and the inside bottle wall, conduction thru the bottle wall, and natural convection between the bottle outer wall and the refrigerator air, in parallel releasing heat from the bottle’s outer wall as radiation.  See Figure 1 and Figure 2.

The assumptions in the thermal resistance model are (1) one-dimensional heat transfer in the radial direction, (2) heat transfer from the top and bottom of the bottle is negligible, (3) the bottle label is ignored.  These assumptions can be included in a more detailed analysis, but for our purposes are sufficient.

The thermal resistance for conduction in Cartesian coordinates is

$$
R_{cond,r}=\frac{L}{(k A_s)}
$$

where L is the thickness.  In our case for cylindrical coordinates, the equation for conduction is

$$
R_{cond,r}=\frac{\ln{(\frac{r_2}{r_1}})}{(2 \pi k L)}
$$

Radial Convection is given by

$$
R_{conv}=\frac{1}{h A_s}=\frac{1}{(h*2 \pi r H)}
$$

Where *H* is the height of the bottle. Finally radiation is given by

$$
R_{rad}=\frac{1}{(h_r*2 \pi r H)}
$$

For our problem, the resistance can become

$$
\begin{align}
& R_{1,conv}=\frac{1}{h_{i}*2\pi r H_{eqiv}}  \\
& R_{2,cond}=\frac{\ln(\frac{r_{out}}{r_{in}})}{2\pi k L}  \\
& R_{3,conv}=\frac{1}{h_{o}*2\pi r H_{eqiv}}  \\
& R_{4,rad}=\frac{1}{h_{r}*2\pi r H_{eqiv}}  \\
& H_{eqiv} = \frac{V_L}{\pi r_{i}^2}
\end{align}
$$

The total resistance will add resistance in series and parallel, to yeild

$$
U = \frac{1}{R_{total}} = R_{1,conv}+R_{2,cond}+ \
(\frac{1}{R_{3,conv}}+\frac{1}{R_{4,rad}})^{-1}
$$

The governing equation will therefore be

$$
\frac{\Delta{T}}{R_{total}}=\frac{(T_{center}-T_{2,\infty})}{R_{total}}=\rho V c \frac{d\mathbf{T}}{d\mathbf{t}}\\
$$

Rearranging

$$
{d\mathbf{t}}\left[\frac{(T_{center}-T_{2,\infty})}{(\rho V c )(R_{total})}\right]= d\mathbf{T}
$$

Where the temperature at the center of the bottle will have to be numerically solved.  The next step will be to calculate the thermal resistances by first finding the natural convection and radiation coefficients.

## Natural Convection Coefficients

Now to determine the coefficients of convection for the beverage inside the bottle $h_i$, the air outside the bottle $h_o$, and radiation to the environment $h_r$.

To help determine these coefficients, an experiment was performed to collect data of the beverage’s temperature drop versus time from ambient room temperature $T_{atm}$ and the temperature inside the refrigerator $T_{inf}$.

The same temperatures as recorded during the experiment were used to approximate accurate coefficients of convections.

$$
T_{atm}=294 \deg K (69.7 \deg F) \\
T_{inf}=276 \deg K (37.6 \deg F)
$$

More on the experiment and data collection will be discussed later.

## Natural Convection Outside of Bottle, $h_o$

Heat transfer on the outside of the bottle related to R3 occurs from the natural (free) convection generated from the warmer air closer to the bottle rising as the cooler air fills the void.

From Eq 9.21 from Ref [1], the average convection coefficient can be found by the Nusselt number, a dimensionless parameter of the ratio of the thermal and mass diffusivities.

$$
\overline{Nu}=\frac{\overline{h_o}L}{k_f}=\frac{4}{3}\bigl(\frac{G_{rL}}{4}\bigr)^{(1/4)}*g(Pr)
$$

Where Pr is the Pradult number, interpolated from Table A.4 of empirical data based on $T_{inf}$.

$$
Pr = \frac{(.707-.720)}{(300-250)}*\bigl( T_{inf}-250 \bigr) + .720 \\
$$

The Pradult number is used in the empirically found function below, Eq 9.20 [1]

$$
g(Pr) = \frac{(0.75*Pr)^{(1/2)}}{(0.609+1.221Pr^{1/2}+1.238Pr)^{(1/4)}}
$$

Next is the Grahsof Number Gr, a measure of the ratio of the buoyancy forces to the
viscous forces acting on the fluid, Eq 9.12 [1].

$$
GR = \frac{(g \beta *(T_{atm}-T_{inf})*L^3)}{\nu ^2}
$$

Where $\beta$ is Volumetric Thermal Expansion Coefficient, for Air Ideal Gas 

$$\beta = \frac{1}{T_{inf}}$$

L is the length of wall with natural convection cooling, and ν is the dynamic viscosity of air interpolated from Table A.4 [1]

$$
\nu = \Bigl[\frac{(15.89-11.44)}{(300-250)}*(T_{inf}-250)+11.44 \Bigr]*10^{-6}
$$

and gravity

$$
g = 9.81 [m/s^2]
$$

Last but not least the coefficient of conduction for the fluid air $k_f$.  Interpolation from Table A.4 [1] yields

$$
K_f = 0.0244 [W/m*K]
$$

## Natural Convection Inside of Bottle, $h_i$

The internal natural convection coefficient can be found from an another empirically derived equation, found in Ref [3] and used in Ref [2] Thermal Performance of Aluminum and Glass Beer Bottles, Ref Eq 13&14.

The natural convection coefficient at the inner bottle surface can be approximated using the following expression for internal flow within a vertical cylinder [2]

Using fluid properties for water (future work can be done to use carbonated and pressurized beer for better approximations)

$$
\frac{h_i H}{k}= 0.55Ra^{0.25}
$$

Where Ra In fluid mechanics is the Rayleigh number (Ra), a dimensionless number associated with buoyancy-driven flow

$$
Ra = \frac{g \beta (T_s-T_{inf})H^3}{\nu \alpha}
$$
